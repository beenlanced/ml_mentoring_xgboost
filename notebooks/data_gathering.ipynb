### Kaggle heart.csv dataset - Used for Classification of Heart Disease

Understanding Heart Disease

Heart disease,often referred to as cardiovascular disease, encompasses a variety of conditions affecting the heart and blood vessels It remains a leading global health concern.

##### Cmmon Forms

* **Atherosclerosis** – Hardening of arteries
* **Valve disorders** – Valve function issues
* **Cardiomyopathy** – Heart muscle disease
* **Congenital defects** – Problems from birth

This data set explores the cardiac features of patients. 

This dataset provides various information about patients, including age, gender, blood pressure, cholesterol levels, electrocardiographic (ECG) features, and more.

##### Dataset Information:

This dataset includes the following features:

* age: The age of the patient.
* sex: Gender of the patient (0: male, 1: female).
* cp: Type of chest pain.
* trestbps: Resting blood pressure.
* chol: Serum cholesterol.
* fbs: Fasting blood sugar > 120 mg/dl.
* restecg: Resting electrocardiographic results.
* thalach: Maximum heart rate achieved.
* exang: Exercise induced angina.
* oldpeak: ST depression induced by exercise relative to rest
* slope (Dtype: int): Slope of peak exsercis ST
* ca (Dtype: int): Major vessels (0-4) - codl standar predicter
* thal (Dtype: int): Thalassemia 4 types 
    1 `Alpha Thalassemia Minima`: One faulty gene; usually asymptomatic.
    2 `Alpha Thalassemia Minor`: Two faulty genes; may have mild symptoms.
    3 `Hemoglobin H Disease`: Three faulty genes; can cause moderate to severe symptoms.
    4 `Hydrops Fetalis`: Four faulty genes; often results in stillbirth or requires lifelong treatment.

##### One Target Variable for Classification 

* target (Dtype: int64): Heart Disease Status (0: Does not have heart disease, 1: Has heart disease)

### Feature Effects that May Lead to Heart Disease

* age: The risk of heart disease increases > 55
* sex: Higher risk in Men (~60%)
* cp: Angina = 80% Disease chance
* trestbps: Target < 140 mmHg.
* chol: Clogging risk > 200 mg/dl. 
* fbs: Diabestes indicator (>120).
* restecg: Abnormal = 60% Risk.
* thalach: Higher Max HR is healthier
* exang: Exercise Pain = Major Flag.
* oldpeak: Stress-induced ischemia
* slope Flat/Down patterns = Disease 
* ca Gold standard predicter
* thal (Dtype: int): Thalassemia 4 types  (>2 issues grow) 4 is the worst

### Saving the dataset to .csv file

```python
# Data File constants
file_name = "heart.csv"
kaggle_data_path = "arezaei81/heartcsv"
data_file_path = Path("data") / file_name

# # Check if the target file already exists to avoid having to download it
if data_file_path.exists():
    print(f"{data_file_path} already exists. Skip download and extract from Kaggle")
else:
    # Load the latest version and store in .csv for future use
    df = kagglehub.dataset_load(
        KaggleDatasetAdapter.PANDAS,
        "arezaei81/heartcsv",
        file_name,
    )
    # save data to csv for future uses
    df.to_csv(data_file_path, index=False)
```
